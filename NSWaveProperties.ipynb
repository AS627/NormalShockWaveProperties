```python
import numpy as np
import matplotlib.pyplot as plt
gamma = 1.4

ma1 = np.linspace(0.1,4,100)

ma2 = ((ma1**2 + (2/(gamma - 1))) / (((gamma*2)/(gamma - 1))*ma1**2 - 1))**(0.5)
ma2_num = np.linspace(0.1, 4, len(ma2))

p2_p1 = 1 + ((2*gamma)/(1+gamma))*(ma1**2-1)
rho2_rho1 = (1- (2/(gamma + 1))*((ma1**2-1)/ma1**2))**(-1)
T2_T1 = 1 + ( (2*(gamma - 1) /((gamma+1)**2))*((ma1**2-1)/ma1**2)*(gamma*ma1**2 + 1))

a = ma1**2 *(gamma+1)/2
b = 1 + (ma1**2)*(gamma -1) /2
c = (a/b)**((gamma)/(gamma-1))
d = ma1**2 * ((2*gamma)/ (gamma+1))
e = (gamma-1) / (gamma + 1)
f = (d - e) ** (-1 / (gamma-1))
Po2_Po1 = c * f
To2_To1 = np.linspace(1,1, len(ma1))
```

```python

plt.plot(ma2_num,ma2, 'k', label = 'Ma2')
plt.plot(ma1,rho2_rho1, 'r', label = 'rho2_rho1')
plt.plot(ma1, p2_p1, 'b', label = 'p2_p1')
plt.plot(ma1, T2_T1, 'g', label = 'T2_T1')
plt.plot(ma1, To2_To1, 'c', label = 'To2/To1')
plt.plot(ma1, Po2_Po1, 'm', label = 'Po2/Po1')

plt.ylim([0,3])
plt.xlim([0.5,4])
plt.title('Flow Properties across NS wave')
plt.xlabel('Ma1')
plt.ylabel('Ma2')
```

```python
d_sr = np.log(To2_To1)**(gamma/ (gamma-1)) - np.log(Po2_Po1)
plt.plot(ma1, d_sr)
plt.ylim([-2,3])
plt.xlim([0.5, 4])
plt.title('ds/R = F(gamma, ma1)')
plt.xlabel('ma1')
plt.ylabel('ds/R')
```
